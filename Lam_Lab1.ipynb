#### <span style="color:darkblue"> Question 5a. Determine the number of WHO regions for which we have available data on infant deaths each year. </span> 


```r
#5A
#of WHO regions for which we have available data on infant deaths each year 
#
#filter(!is.na(mph)=keep observations that is not NA

library(ggplot2)
library(dplyr)
dta <- read.csv("http://apps.who.int/gho/athena/data/xmart.csv?target=GHO/MORT_200&profile=crosstable&filter=MGHEREG:WORLD;MGHEREG:REG6_AFR;MGHEREG:REG6_AMR;MGHEREG:REG6_EMR;MGHEREG:REG6_EUR;MGHEREG:REG6_SEAR;MGHEREG:REG6_WPR;CHILDCAUSE:CH3&x-sideaxis=MGHEREG;YEAR&x-topaxis=GHO;CHILDCAUSE;AGEGROUP",
               skip=2)
who.DD <- data.frame(Year = dta$Year, WHO.region = dta$Region, 
                        infantDeaths.0.27days = dta$X.0.27.days,
                        infantDeaths.1.59mos = dta$X.1.59.months,
                        infantDeaths.0.4yrs = dta$X.0.4.years)
head(who.DD)
dim(who.DD)
AVAILABLEDATA<-filter(who.DD,!is.na(infantDeaths.0.27days),
                      !is.na(infantDeaths.1.59mos),
       !is.na(infantDeaths.0.4yrs))

summarise(AVAILABLEDATA,n())
dim(AVAILABLEDATA)
#AVAILABLEDATA%>% group_by(WHO.region)
tally(AVAILABLEDATA%>% group_by(Year))
```

#### <span style="color:darkblue"> Question 5c: Construct side-by-side boxplots to illustrate and compare the distributions of infant deaths that occured (per 1000 live births) in 2015 for the first two age ranges (0-27 days and 1-59 months). What can you conclude from this visualization?

```r
library(tidyr)
whoNew <- who.DD %>%
gather(age, infantDeaths,
c(infantDeaths.0.27days,infantDeaths.1.59mos))

Data2015<-filter(whoNew,Year==2015)
Data2015

ggplot(Data2015, 
    aes(x=age,y=infantDeaths,fill=age)) +
    geom_boxplot(alpha=.8)+
    scale_fill_brewer(palette="Accent") +
    theme(axis.text.x = element_blank())+ggtitle("Infant Deaths in 2015")
```

#### Solution in words goes here

From this visualization, we can see that infants who are 1 - 59 months old have larger variability in the number of deaths than that of infants who are 27 days old. The median number of infant deaths for babies who are 27 days old is 0.1 per 1000 live births, while the median rate of deaths for infants is 3.6 deaths per 1000 live births. The distribution of infants who are 1-59 months old is skewed to the right, while the distribution for infants who are 27 days seems approximately normal, but it's difficult to tell because the boxplot looks so compressed.
